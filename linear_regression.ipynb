```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#DATA to be plotted
heights = [[1.6],[1.65],[1.7],[1.73],[1.8]]

weights = [[60],[65],[72.3],[75],[80]]
```

```python
#plotting
plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters')
plt.ylabel('Weights in kilograms')
plt.plot(heights, weights, 'k.')
plt.plot(heights, model.predict(heights), 'r')

#axis range for x and y
plt.axis = ([1.5,1.85,50,90])
plt.grid(True)
```

```python
#Using Linear Regression class
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X=heights, y=weights)
```

```python
#making predictions
weight = model.predict([[1.75]])[0][0]
print(round(weight,2))
```

```python
model.intercept_[0]
```

```python
model.coef_[0][0]
```

```python
#Examining performance using the Residual Sum of Squares
import numpy as np
print('Residual sum of squares: %.2f'%np.sum((weights - model.predict(heights))**2))
```

```python
#testing : this is done by measuring how closely the test data fits the regression line
# The way to do this is using the R-squared method (coefficient of determination)
heights_test = [[1.58],[1.62],[1.69],[1.76],[1.82]]
weights_test = [[58],[63],[72],[73],[85]]

r_squared = model.score(heights_test, weights_test)
print(round(r_squared,4))

```

```python
#Saving the model(Persisting the model)
import pickle
filename = 'h_w_model.pkl'
pickle.dump(model, open(filename,'wb'))
```

```python
#using the model
loaded_model = pickle.load(open(filename, 'rb'))
```
